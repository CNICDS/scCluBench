```python
colors = ['#589EA6', '#F28F38']
plt.figure(figsize=(8, 6))
bins = np.linspace(0, 30000, 7)
width = 4000
plt.hist([mouse_num_cells+500, human_num_cells+500], bins=bins+500, stacked=True, label=['Mouse', 'Human'], width=width, color=colors)
plt.xticks([0, 5000, 10000, 15000, 20000, 25000, 30000], ['0', '5K', '10K', '15K', '20K', '25K', '30K'], fontsize=24)
# plt.xlabel('Number of Cells', fontsize=14)
plt.yticks([0, 5, 10, 15, 20, 25], fontsize=24)
plt.legend(fontsize=30)
plt.savefig('dataset/num_cells_histogram.pdf', bbox_inches='tight')
plt.close()
plt.figure(figsize=(8, 6))
bins = np.linspace(0, 70000, 8)
width = 8000
plt.hist([mouse_num_genes+1000, human_num_genes+1000], bins=bins+1000, stacked=True, label=['Mouse','Human'], width=width, color=colors)
plt.xticks([0, 10000, 20000, 30000, 40000, 50000, 60000, 70000], ['0', '10K', '20K', '30K', '40K', '50K', '60K', '70K'], fontsize=24)
# plt.xlabel('Number of Genes', fontsize=14)
plt.yticks([0, 5, 10, 15, 20], fontsize=24)
plt.legend(fontsize=30)
plt.savefig('dataset/num_genes_histogram.pdf', bbox_inches='tight')
plt.close()
plt.figure(figsize=(8, 6))
bins = np.linspace(0, 40, 9)
width = 4
plt.hist([mouse_num_clusters+0.5, human_num_clusters+0.5], bins=bins+0.5, stacked=True, label=['Mouse', 'Human'], width=width, color=colors)
plt.xticks([0, 5, 10, 15, 20, 25, 30, 35, 40], ['0', '5', '10', '15', '20', '25', '30', '35', '40'], fontsize=24)
# plt.xlabel('Number of Clusters', fontsize=14)
plt.yticks([0, 5, 10, 15], fontsize=24)
plt.ylim(0, 16)
plt.legend(fontsize=30)
plt.savefig('dataset/num_clusters_histogram.pdf', bbox_inches='tight')
plt.close()
plt.figure(figsize=(8, 6))
bins = np.linspace(65, 100, 8)
width = 4
plt.hist([mouse_sparsity+0.5, human_sparsity+0.5], bins=bins+0.5, stacked=True, label=['Mouse', 'Human'], width=width, color=colors)
plt.xticks([65, 70, 75, 80, 85, 90, 95, 100], ['65', '70', '75', '80', '85', '90', '95', '100'], fontsize=24)
# plt.xlabel('Sparsity(%)', fontsize=14)
plt.yticks([0, 5, 10, 15, 20, 25], fontsize=24)
plt.ylim(0, 26)
plt.legend(fontsize=30, loc='upper left')
plt.savefig('dataset/sparsity_histogram.pdf', bbox_inches='tight')
plt.close()
```
